```python
from pylab import *
x=[2,3,4,5,7]
y=[4,2,3,7,9]
if len(x) !=len(y):
    print("number of values of x and y doesnot match")
else:
    scatter(x,y)
    plot(x,y)
    grid()
    xlabel("x-axis")
    ylabel("y-axis")
    title("curve connecting the given points")
    axvline(x=0,color="red")
    axhline(y=0,color="red")
    show()
```

```python
#Program to plot exponential function y = e^x
from pylab import *
x = arange(-5,5,0.001)
y = exp(x)
plot(x,y)
title("$ y=e^x $")
grid()
xlabel("x-axis")
ylabel("y-axis")
xlim(-5,5)
ylim(-1,20)
axvline(x=0, color = "red")
axhline(y=0, color = "red")
show()
```

```python
#Program to plot sin(x) and cos(x) together
from pylab import *
x = linspace(-2*pi,2*pi,100)
plot(x,sin(x),label = "$y=sin(x)$")
plot(x,cos(x),label="$y=cos(x)$")
title("exponential curve")
grid()
xlabel("x-axis")
ylabel("y-axis")
title("sin(x) and cos(x)")
legend()
xlim(-2*pi,2*pi)
axvline(color="red")
axhline(color="red")
show()
```

```python
#Program to plot parametric curve,circle x=rcos(theta) and y=rsin(theta)
from pylab import *
theta = linspace(0,2*pi,1000)
x=3*cos(theta)
y=3*sin(theta)
plot(x,y)
axis("square")
title("$x=3 cos(\\theta)$  &  $ y=3 sin(\\theta)$")
axvline(color="red")
axhline(color="red")
grid()
show()
```

```python
#Program to plot circle r = 5
from pylab import *
theta = linspace(0,2*pi,1000)
r = [5 for i in theta]
polar(theta,r,color = "red")
title("$circle: r = 5 $")
axvline (color = "green")
show()
```

```python
#Program to plot rose r=2|cos(2theta)|
from pylab import *
theta = linspace(0,2*pi,1000)
polar(theta,r)
title("$r = 2|cos(2*\\theta)|$")
axvline(color = "red")
show()
```

```python
#Program to plot cardioids r = a(1+cos(theta)) and r = a(1-cos(theta)) taking a=2
from pylab import *
theta = linspace(0,2*pi,1000)
r1 = 2*(1+cos(theta))
r2 = 2*(1-cos(theta))
polar(theta,r1,theta,r2)
title("cardiods : $r=a(1+cos(\\theta))$ & r=a(1-cos(\\theta))$")
axvline(color="red")
show()
```

```python
from pylab import *
from sympy import *
x,y=symbols('x,y')
p1=plot_implicit(Eq(x**2+y**2,4),xlabel = False,ylabel = False,title = 'Circle : $x^2+y^2=4$ ',aspect_ratio=(2.,2.))
show()
```
